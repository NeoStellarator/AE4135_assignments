```python
#Imports
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

from Rotor import Rotor
import multi_rotor_analysis
import plotting_routines
from globals import plot_dir, res_dir, data_dir

U0 = 60
rho  = 1.007
D = 0.7*2
```

## Calculating area averaged axial induction a

```python
rotor = Rotor(
    # Geometry definition
    c_R_func  = lambda r_R : 0.2455-0.0305*r_R,
    twst_func = lambda r_R : -43.63*r_R+60.87,
    pitch = 0,
    B     = 6,
    R     = 0.7,
    r_R_H = 0.25,
    polar_path = data_dir.joinpath('ARAD8pct_polar.txt'),
    # Operating condition
    J = 60/(20*1.4),
    Vinf = 60,
    rho = 1.067,
    # Discretizaiton
    n_elem = 100,
    dist_elem =  'uniform',
    isPropeller = True
    )

annuli_a = [an.a for an in rotor.annuli_lst]
a_mean = sum(annuli_a*rotor.annuli_A_R2*rotor.R**2)/(np.pi*rotor.R**2)
print(f"a_mean: {a_mean:.3f}")
rotor.plot_check()
```

## Calculating AD Ct and Cp

```python
print(f"a_mean: {a_mean:.3f}")
TC_bem = rotor.TC
TC_ad = 4*a_mean*(1+a_mean)

PC_bem = rotor.PC
PC_ad = 4*a_mean*(1+a_mean)**2

print()

print(f"Tc_bem = {TC_bem}, Tc_ad = {TC_ad}")
print(f"Pc_bem = {PC_bem}, Pc_ad = {PC_ad}")

```
