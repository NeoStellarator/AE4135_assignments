```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(-1, 3, 100)
CT = np.zeros(a.shape)

CT1 = 1.816
ac = 1-np.sqrt(CT1)/2
CTc = 4*ac*(1-ac)
CT[a<ac]  = 4*a[a<ac]*(1-a[a<ac])
CT[a>=ac] = CT1-4*(np.sqrt(CT1)-1)*(1-a[a>=ac])

CT_  = 4*a*(1-a)

fig, ax = plt.subplots()

ax.plot(a, CT)
ax.plot(a, CT_, ls='--')
ax.plot([ac], [CTc], 'ro')
```

```python
ac, CTc
```

```python
## propeller

a = np.linspace(-1, 3, 100)
CT = np.zeros(a.shape)


CT1 = -1.818
ac = np.sqrt(np.abs(CT1))/2-1
CTc = 4*ac*(1+ac)

CT[a>ac] = 4*a[a>ac]*(1+a[a>ac])
CT[a<ac] = CT1 + 4*(np.sqrt(np.abs(CT1))-1)*(1+a[a<ac])

CT_  = 4*a*(1+a)

fig, ax = plt.subplots()

ax.plot(a, CT)
ax.plot(a, CT_, ls='--')
ax.plot([ac], [CTc], 'ro')
ac, CTc
```
